como argumento predefinido en 150 la funcion debe de regresar el pago del empleado en un
dia, en una semana (5 dıas) y en un mes (20 dıas).


```
def pago_empleado (horas_trabajando, costo_hora=150):
    undia = horas_trabajando*costo_hora
    unasemana = undia*5
    unmes = unasemana*4
    pago = [undia,unasemana,unmes]
    return pago
    

a = float(input("Horas Trabajando al dia: "))

print("Los pagos son: \n", pago_empleado(a))

    
```

9.- Escribir una funcion que reciba un numero entero positivo y devuelva su factorial

```
def factorial(n): 
    resultado = 1
    i = 1
    while i <= n:
        resultado = resultado * i
        i = i + 1
    return resultado

n = int(input("Dame un valor: "))
print("El factorial es: ", factorial(n))
```
